# Morphological transformation

Date: 2021/8/29

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def process(filename):
    src = cv.imread(filename)
    src = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
    
    kernel = np.ones((7, 7), np.uint8)
    opening = cv.morphologyEx(src, cv.MORPH_OPEN, kernel)

    dst = cv.bitwise_not(opening)
    
    src = cv.cvtColor(src, cv.COLOR_BGR2RGB)
    opening= cv.cvtColor(opening, cv.COLOR_BGR2RGB)
    dst = cv.cvtColor(dst, cv.COLOR_BGR2RGB)

    fig = plt.figure(1, (15., 15.))
    plt.imshow(cv.vconcat([src, opening, dst]))
```

```python
process('sample4.jpg')
```
