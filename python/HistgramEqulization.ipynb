# Histgram equalization

Date: 2021/8/29

```python
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def process(filename):
    mat = cv.imread(filename)
    mat2 = cv.cvtColor(mat, cv.COLOR_BGR2GRAY)
    mat2 = cv.equalizeHist(mat2)

    img = cv.cvtColor(mat, cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(mat2, cv.COLOR_GRAY2RGB)
    fig = plt.figure(1, (10., 10.))
    plt.imshow(cv.hconcat([img, img2]))
```

```python
process('sample1.jpg')
```
