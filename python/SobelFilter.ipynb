# Scratch detection with Sobel filter

Date: 2021/8/29

```python
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def process(filename, xy='x', ksize=3):
    src = cv.imread(filename)
    gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
    blur = cv.GaussianBlur(gray, (3, 3), 0)
    src = src[0:3000, 0:2000]
    dst = blur[0:3000, 0:2000]

    if (xy == 'x'):
        dst = cv.Sobel(dst, cv.CV_8UC1, 1, 0, ksize=ksize)
    elif (xy == 'y'):
        dst = cv.Sobel(dst, cv.CV_8UC1, 0, 1, ksize=ksize)
    
    img = cv.cvtColor(dst, cv.COLOR_BGR2RGB)
    fig = plt.figure(1, (12., 12.))
    plt.imshow(cv.hconcat([cv.cvtColor(src, cv.COLOR_BGR2RGB), img]))
```

```python
process('sample3.jpg', 'x', 5)
```

```python
process('sample3.jpg', 'y', 5)
```
